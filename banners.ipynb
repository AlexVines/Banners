```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt
```

```python
# Data source: https://archive.org/details/201309_foursquare_dataset_umn
df = pd.read_table("checkins.dat", sep="|", skipinitialspace=True, low_memory=False)
df.rename(columns=lambda x: x.strip(), inplace=True)
df = df[['latitude', 'longitude']].dropna()[:100000]
df.head(10)
```

```python
plt.scatter(df['longitude'], df['latitude'], s=10)
```

```python
#Get coordinates from Google Maps
offices_coords = np.array([[33.751277, 25.867736, 51.503016, 52.378894, 39.366487, -33.868457],
                           [-118.188740, -80.324116, -0.075479, 4.885084, 117.036146, 151.205134]])
offices = offices_coords.T
```

```python
ms = MeanShift(bandwidth=0.1)
ms.fit(df)
```

```python
vals, count = np.unique(ms.labels_, return_counts=True)
len(vals)
```

```python
#Get labels which have 15 or more elements in it
filtered = [[i,j] for i,j in zip(vals, count) if j>15]
fit_labels, counts = a,b = list(zip(*filtered))
```

```python
cluster_centers = ms.cluster_centers_
fit_coords = []
for i in range(len(cluster_centers)):
    if i in fit_labels:
        fit_coords.append(cluster_centers[i])
len(fit_coords)
```

```python
def dist(A, B):
    return np.sqrt((A[0]-B[0])**2 + (A[1]-B[1])**2)

distanses = []
for i in range(len(offices)):
    for j in range(len(fit_coords)):
        empt = []
        empt.append(dist(offices[i], fit_coords[j]))
        empt.append(fit_coords[j])
        distanses.append(empt)

distanses.sort()
distanses = distanses[:20]
distanses
#Here we can see coordinates of 20 closest places for banners
```

```python
# a, b = list(zip(*distanses))
# c = list(b)
plt.figure(figsize=(20,10))
x = [row[1] for row in offices]
y = [row[0] for row in offices]
plt.scatter(x, y, s=50)
plt.scatter(df['longitude'], df['latitude'], s=10, c='g')
plt.scatter(x, y, c='r', s=50)
```
